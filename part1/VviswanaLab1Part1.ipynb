```r
# Problem 1: Define two synthetic vectors of data representing sales over 12 months for 2 items. Compare the two using lines graphs.

sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34)  #12 numbers with a mean of 34
par(bg="cornsilk")
plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" ) # X-axis is Month and Y -axis is Sales
title(main="Sales by Month") # The title of the Visualisation
lines(sales2, type="o", pch=22, lty=2, col="red") # The second plot in the same graph
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```

```r
# Problem 6: Accessing external APIs: eg. Google map lat-long API: map command. Get API key from Google Cloud.
# The idea here is to plot the results of analysis on a map: geographical or otherwise. List a collection of cities you have visited and plot it on a map.

#Importing the necessary packages
library("ggmap") 
library("maptools")
library(maps)
#The google API key is registered and its geolocation is turned on.
register_google(key = "Google API Key", write = TRUE) 
#The visited cities are entered into a vector
visited <- c("SFO", "Chennai", "London", "Melbourne", "Lima, Peru", "Johannesbury, SA")
ll.visited<- geocode(visited)
# The latitude and longitude of the visited cities are taken
visit.x<- ll.visited$lon
visit.y<- ll.visited$lat
#The background color of the world map is given as lightblue
map("world", fill=TRUE, col="white", bg="lightblue", ylim=c(-60, 90), mar=c(0,0,0,0))
# The lat and lon of the visited cities are marked in red
points(visit.x,visit.y, col="red", pch=36)

#Here is another example using the map of The United States.
#Follows the same process as above except the fact here we use a World Map.
library("ggmap")
library("maptools")
library(maps)
visited<- c("SFO", "New York", "Buffalo", "Dallas, TX")
ll.visited<- geocode(visited)
visit.x<- ll.visited$lon
visit.y<- ll.visited$lat
map("state", fill=TRUE, col=rainbow(50), bg="lightblue", mar=c(0,0,0,0))
points(visit.x,visit.y, col="yellow", pch=36)
```
